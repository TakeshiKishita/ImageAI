```python
from imageai.Prediction.Custom import CustomImagePrediction
import os
from IPython.display import Image, display_jpeg
```

### 画像の確認
affenpinscherディレクトリに入っている画像なので、この画像の犬種は「affenpinscher」が正しい

```python
jpg_file = "Images/test/affenpinscher/n02110627_8621.jpg"
display_jpeg(Image(jpg_file))
```

### 画像の推論

```python
execution_path = os.getcwd()

prediction = CustomImagePrediction()
prediction.setModelTypeAsResNet()
prediction.setModelPath(os.path.join(execution_path, "Images/models/model_ex-016_acc-0.303571.h5"))
prediction.setJsonPath(os.path.join(execution_path, "Images/json/model_class.json"))
prediction.loadModel(num_objects=120)  # 

predictions, probabilities = prediction.predictImage(os.path.join(execution_path, jpg_file), result_count=5)

for eachPrediction, eachProbability in zip(predictions, probabilities):
    print(eachPrediction , " : " , eachProbability)
```
